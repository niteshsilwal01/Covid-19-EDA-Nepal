## Feature Engineering

```python
#converting string date to date-time
import datetime as dt
covid_nepal['date'] = pd.to_datetime(covid_nepal['date']) 
covid_nepal.head()
```

```python
#converting date-time to ordinal
covid_nepal['date']=covid_nepal['date'].map(dt.datetime.toordinal)
covid_nepal.head()
```

## Modelling -01 : Total Cases 

Let's build a regression model to make the prediction of the total cases in Nepal in the coming days if no precautions taken.

```python
#Data Split
X=covid_nepal['date']
y=covid_nepal['total_cases']
```

```python
#Train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)      
```

Linear Regression

```python
#importing and fitting the model to generate the pattern.
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression()
lin_model.fit(np.array(X_train).reshape(-1,1),np.array(y_train).reshape(-1,1)) #reshaping train data as it consists of 1-d data. 
```

```python
#Using the generated pattern to predict the total cases.
y_predicted=lin_model.predict(np.array(X_test).reshape(-1,1))
```

```python
#Evaluation metrics
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_predicted)
print(f'MSE is {round(mse)}')

score=lin_model.score(np.array(X_test).reshape(-1,1),np.array(y_test).reshape(-1,1))
print(f'Model score is :{(score)}')
```
